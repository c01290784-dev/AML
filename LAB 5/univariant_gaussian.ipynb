```python
#create a data set
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

data = [
4.72, 4.89, 5.01, 5.12, 5.18,
5.24, 5.31, 5.38, 5.44, 5.51,
5.57, 5.63, 5.69, 5.74, 5.79,
5.83, 5.88, 5.92, 5.96, 5.99,
6.02, 6.05, 6.08, 6.11, 6.14,
6.17, 6.20, 6.23, 6.27, 6.31,
6.35, 6.39, 6.44, 6.49, 6.54,
6.60, 6.66, 6.73, 6.80, 6.88,
6.96, 7.05, 7.14, 7.24, 7.35,
5.47, 5.76, 6.04, 6.32, 6.58
]
```

```python
#step1: calculate the sample mean and variance of the data
mean=np.mean(data)
variance=np.var(data)

print("mean:",mean)
print("variance:",variance)
```

```python
x=np.linspace(min(data)-1,max(data)+1,100)
pdf= norm.pdf(x,mean,np.sqrt(variance))

plt.hist(data,bins=8,density=True , alpha=0.6, label="data histogram")
plt.plot(x,pdf,'r',label="gaussia pdf")
plt.title("univariate gaussian distribution ")
plt.xlabel("electricity usage (kwh)")
plt.ylabel("density")
plt.legend()
plt.show()
```

```python
#task2 
#understand covaraince and elliptical distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

np. random.seed(42)

#generate data
daily=np.random.normal(5,0.5,100)
peak=np.random.normal(2,0.3,100)

X=np.column_stack((daily,peak))

#estimate parameters
mean_vector = np.mean(X,axis=0)
cov_matrix= np.cov(X.T)

print("mean vector:\n", mean_vector)
print("convariance matrix:\n", cov_matrix)

#create gaussian model
rv=multivariate_normal(mean_vector,cov_matrix)

#create grid
x,y= np.mgrid[3:7:0.05,1:3:0.05]
pos=np.dstack((x,y))


#plot contour
plt.figure(figsize=(6,5))
plt.contour(x,y,rv.pdf(pos),levels=20,cmap='viridis')
plt.colorbar(label='probability density')

#plot data
plt.scatter(X[:,0],X[:,1],c='red',edgecolors='k', label='data points')

plt.title("multivariate gaussian distribution ")
plt.xlabel("daily usage ")
plt.ylabel("peak usage")
plt.legend()
plt.show()
```
